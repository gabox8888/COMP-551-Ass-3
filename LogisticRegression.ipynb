```python
import numpy as np
import pandas as pd
import csv


trainLabel = []

with open('train_y.csv') as csvfile:
    trainReader = csv.reader(csvfile)
    for i, row in enumerate(trainReader):
        if i != 0:
            try:
                for j, col in enumerate(row):
                    if j !=0:
                        trainLabel.append(int(col))
            except :
                print(i)
                    

label = np.zeros((100000,19))

for i,output in enumerate(trainLabel):
    label[i,output]=1
                
                
np.savetxt("output.csv",label, delimiter=",")   
```

```python
#Linear regression for the wavelet on small image
import random

trainFactor = 0.005
nbEpoch = 25

wavTrain = np.genfromtxt('waveletTrainSmall.csv', delimiter=',')

wavShape = wavTrain.shape

w = np.random.rand(wavShape[1],19)

rand = np.random.rand(wavShape[0])

train = np.array(wavTrain[rand<0.7,:])
test = np.array(wavTrain[0.9<=rand,:])
validation = np.squeeze(np.array(wavTrain[(np.logical_and(rand<0.9 , rand>=0.7)),:]))


trainLabel = np.array(label[rand<0.7,:])
testLabel = np.array(label[0.9<=rand,:])
validationLabel = np.squeeze(np.array(label[(np.logical_and(rand<0.9 , rand>=0.7)),:]))

trainShape = train.shape
valShape = validation.shape
testShape = test.shape

trainResultSmallWavelet = []
valResultSmallWavelet = []

predict = np.zeros((19,1))

for i in range(0,nbEpoch):
    
    correctAnswer = 0
    for j in range(0,trainShape[0]):
        for k in range(0,19):
            predict[k] = 1/(1+np.exp(-np.sum(w[:,k]*train[j,:])))
        if np.argmax(predict) == np.argmax(trainLabel[j,:]):
            correctAnswer = correctAnswer+1
    
    trainResultSmallWavelet.append(correctAnswer/trainShape[0])
    
    correctAnswer = 0
    for j in range(0,valShape[0]):
        for k in range(0,19):
            predict[k] = 1/(1+np.exp(-np.sum(w[:,k]*validation[j,:])))
        if np.argmax(predict) == np.argmax(validationLabel[j,:]):
            correctAnswer = correctAnswer+1

    valResultSmallWavelet.append(correctAnswer/valShape[0])
    
    for j in range(0,trainShape[0]):
        for k in range(0,19):
            w[:,k] = w[:,k] + trainFactor*train[j,:]*(trainLabel[j,k]-1/(1+np.exp(-np.sum(w[:,k]*train[j,:]))))
    
    
correctAnswer = 0
for j in range(0,testShape[0]):
    for k in range(0,19):
        predict[k] = 1/(1+np.exp(-np.sum(w[:,k]*test[j,:])))
    if np.argmax(predict) == np.argmax(testLabel[j,:]):
        correctAnswer = correctAnswer+1

print("precision test small wavelet = ", correctAnswer/testShape[0])    
```
